```python
#importar bibliotecas
import pandas as pd
import seaborn as srn
import statistics  as sts
```

```python
#importar dados
tabela = pd.read_csv("Tempo.csv", sep=";")
#visulizar
tabela.head()
```

```python
#tamanho
tabela.shape
```

```python
#verificar se há valores NULOS (NAN)
tabela.isnull().sum()
```

```python
#explorar colunas numéricas
tabela['Umidade'].describe()
```

```python
#pegar MEDIANA de Umidade
medianaUmidade = sts.median(tabela['Umidade'])
medianaUmidade
```

```python
#substituir NAN por MEDIANA
tabela['Umidade'].fillna(medianaUmidade, inplace=True)
```

```python
#Verificamos se NAN não existem mais
tabela['Umidade'].isnull().sum()
```

```python
#explorar colunas categoricas
agrupadoVento = tabela.groupby(['Vento']).size()
agrupadoVento
```

```python
#preenche NAN com FALSO (moda -> valor que mais aparece)
tabela['Umidade'].fillna('Falso', inplace=True)
```
