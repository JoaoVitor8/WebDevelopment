```python
# Importações
import pandas as pd
import numpy as np
```

```python
# Dados
loja_df = pd.read_csv("vendas.csv")
display(loja_df)
```

```python
# preencher valores vazios
loja_df['Desconto (%)'] = loja_df['Desconto (%)'].fillna(0)
display(loja_df)
```

```python
# Correção dos dados
loja_df["Categoria"] = loja_df["Categoria"].replace(["Eletronicos", "Eletrônicos"], "Eletro")
display(loja_df)
```

```python
# Adicionando a coluna lucro
loja_df['Lucro'] = loja_df['Total_Venda'] - (loja_df['Total_Venda'] * (loja_df['Desconto (%)'] / 100))
display(loja_df)
```

```python
# Vendas de uma categoria específica, nesse caso Moda
vendas = loja_df.loc[loja_df['Categoria'] == 'Moda']
display(vendas)
```

```python
# Total de vendas por categoria.
loja_soma = loja_df.groupby('Categoria').agg('Total_Venda').sum()
display(loja_soma)
```

```python
# Produtos com maior lucro em ordem decrescente.
loja_produto = loja_df.groupby('Produto').agg('Total_Venda').sum().sort_values(ascending=False)
display(loja_produto)
```

```python
# Data com maior volume de vendas.
loja_data = loja_df.groupby('Data').agg('Quantidade').max().sort_values(ascending=False).head(20)
display(loja_data)
```

```python
# Preço médio por unidade dos produtos.
preco_medio = np.sum(loja_df["Total_Venda"]) / np.sum(loja_df["Quantidade"])
display(f"Preço médio por unidade: R$ {preco_medio:.2f}")
```

```python
# Função que classifica os descontos como Alto (>=20%) ou Baixo (<20%)
def classificar_desconto(desconto):
  return "Alto" if desconto >= 20 else "Baixo"

loja_df["Classificação_Desconto"] = np.vectorize(classificar_desconto)(loja_df["Desconto (%)"])

display(loja_df[["Desconto (%)", "Classificação_Desconto"]])
```

```python
# Calculando a receita total
receita_total = loja_df["Lucro"].sum()
display(f"A receita total dessa loja é de R$ {receita_total:.2f}")
```

```python
# Calculando o lucro médio por venda
lucro_medio = loja_df["Lucro"].mean()
display(f"O lucro medio por venda dessa loja é de R$ {lucro_medio:.2f}")
```

```python
# Variação na quantidade comprada
variacao_quantidade = loja_df["Quantidade"].std()
display(f"Variação na quantidade comprada (desvio padrão): {variacao_quantidade:.2f}")
```

```python
# Número de compras dos clientes
loja_cliente = loja_df.groupby('Cliente_ID').agg('Quantidade').sum().sort_values(ascending=False)
display(loja_cliente)
```

```python
# Produto menos ventidos
produtos_mais_vendito = loja_df.groupby('Produto').agg('Quantidade').sum().sort_values(ascending=False).head(5)
display(produtos_mais_vendito)
```

```python
# Produto menos ventidos
produtos_menos_vendito = loja_df.groupby('Produto').agg('Quantidade').sum().sort_values().head(10)
display(produtos_menos_vendito)
```

```python
# Exportando o dataset limpo
loja_df.to_csv('vendas_atualizado.csv', index=False)
```
